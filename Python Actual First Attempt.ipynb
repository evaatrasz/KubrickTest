Required Signature:

````python
# submit (we will import your solution from this cell. Don't use Global Variables here.  )
# ------------------------------------------------------------------------------

def keep_only_int(x):
    new_list = []
    for item in x:
        if type(item) == int:
            new_list.append(item)
        else:
            continue
    return new_list
        

#keep_only_int([1, 1.5, 2, 'a'])
````

-------------------------------

        It should continue to yield this string for `num_tosses` iterations, and then stop. Each iteration, there should 
        be a 50% chance of yielding `heads`, and a 50% chance of yielding `tails`.   The function should accept

Required Signature:

        pass  

For example, if we wrote `x = coin_toss(2)`, then we could iterate over `x` exactly two times, at each iteration it would yield either `heads` or `tails`.



````python
# submit (we will import your solution from this cell. Don't use Global Variables here.  )
# ------------------------------------------------------------------------------

import random
def coin_toss(noise_prob):
    for i in range(noise_prob):
        result = random.randint(1,2)
        if result == 1:
            print('Heads')
        else:
            print('Tails')

#x = coin_toss(6)
````

6. Name Processing (4)
-------------------------------
Write a function that accepts an input argument list of `full_name` strings 
and returns a dictionary with four keys, `'first'`, `'last'`, `'both'` and `'neither'`. 
The corresponding value for each of these keys should be a list.
The elements of the list input should be added to exactly one of these lists as follows:
- If just the first name starts with a `'z'`, then add to list associated with the `'first'` key.
- If just the last name starts with a `'z'`, then add to list associated with the `'last'` key.
- If both names start with a `'z'`, then add to list associated with the `'both'` key.
- Otherwise, add the name to the list associated with the `'neither'` key.

You may assume that full_name is in lower case, 
and has exactly one space character that separates first and last names.



Required Signature:

    def name_processing4(list_of_names):  
        pass  


for example, using the following  list as input:

`['bjorn borg', 'robert zimmerman', 'zac goldsmith', 'zara tindall'] `

the function would return :

```
{ 'both': [], 
 'neither': ['bjorn borg'], 
 'first': ['zac goldsmith', 'zara tindall'], 
 'last': ['robert zimmerman']}




````python
# submit (we will import your solution from this cell. Don't use Global Variables here.  )
# ------------------------------------------------------------------------------

def name_processing4(list_of_names): 
    dictionary = {}
    first_name_list = []
    both_list = []
    last_name_list = []
    neither_list = []
    for name in list_of_names:
        full_name = name.split()
        if full_name[0].find('z') == 0 and full_name[1][0] != 'z':
            first_name_list.append(name)
            
        elif full_name[1].find('z') == 0 and full_name[0].find('z') == 0:
            both_list.append(name)
        
        elif full_name[1].find('z') == 0:
            last_name_list.append(name)
        
        else:
            neither_list.append(name)
    
     
    dictionary.update({'both':both_list, 
                      'neither':neither_list, 'first':first_name_list, 'last':last_name_list})
    return dictionary
            
    
#name_processing4(['bjorn borg', 'robert zimmerman', 'zac goldsmith', 'zara tindall', 'mr zebra', 'zip zap'])           


````

````python
# submit (we will import your solution from this cell. Don't use Global Variables here.  )
# ------------------------------------------------------------------------------

class GradeStore:
    def __init__(self, mark):
        self.mark = mark
    
    def get_grade(self):
        if self.mark >=70 and self.mark <= 100:
            print('A')
        elif self.mark >= 60 and self.mark <= 69:
            print('B')
        elif self.mark >=50 and self.mark <=59:
            print('C')
        elif self.mark >=0 and self.mark <= 49:
            print('F')
        else:
            print('None')
    


````

`col_names` and rows with explicit index corresponding to `row_indices`.


Required Signature:

    def dataframe_selector2(df,  row_indices, col_names):  
        pass  

Example:
If this is the input dataframe df:

|   | a |   b  |   c | 
| --- | --- | --- | --- | 
| 0 | 1 |  10 |  100 | 
| 1 |  2 |  20 |  200 |

````python
# submit (we will import your solution from this cell. Don't use Global Variables here.  )
# ------------------------------------------------------------------------------
import pandas as pd
def dataframe_selector2(df,  row_indices, col_names):
    df = df.loc[row_indices, col_names]
    return df

````

10. Dataframe Process Data (1)
-------------------------------
Given an input dataframe in which each row holds a student's mark for a specific
question, e.g.

|   | student | question |      topic | mark |
| --- | --- | --- | --- | --- |
| 0 |  Alice  | multiply | arithmetic |    6 |
| 1 |    Bob  | multiply | arithmetic |    3 |
| 2 |  Alice  |  divide  | arithmetic |    5 |
| 3 |    Bob  |  divide  | arithmetic |    4 |
| 4 |  Alice  |  angles  |  geometry  |  10 |
| 5 |    Bob  |  angles  |  geometry  |   5 |

Write a function to add up the total marks awaded to each student 
(assuming each student attempts each question at most one time), and return this in 
a dictionary (4 marks). 

For one further mark, relax this assumption, i.e. only include students' best attempt at each question, to calculate their totals. 

Required Signature:

    def dataframe_process_data1(df):  
        pass  

e.g. for the following input:
`pd.DataFrame({'student':['Alice', 'Bob', 'Alice', 'Bob', 'Alice', 'Bob'], 
            'question':['multiply', 'multiply', 'divide', 'divide', 'angles','angles'],
            'topic':['arithmetic', 'arithmetic', 'arithmetic', 'arithmetic', 'geometry','geometry'],
            'mark':[10,20,30,40,50,60]})`
            
 this is the expected output:

`{'Alice': 90, 'Bob': 120}`



````python
# submit (we will import your solution from this cell. Don't use Global Variables here.  )
# ------------------------------------------------------------------------------

def dataframe_process_data2(df):
    import numpy as np
    new_dataframe = df.groupby('student').agg({'mark':np.sum})
    dictionary = {}
    new_dataframe = new_dataframe.reset_index()
    for value in new_dataframe.values:
        key = value[0] 
        dicvalue = value[1]
        dictionary.update({key:dicvalue})
    return dictionary


````

The restful web service `api.publicapis.org` provides information about other web services that are available.

Write a function that uses its `entries` endpoint to return a list  of the `'Link'` elements for these items. 

Your function should accept a single argument, `category`, 
which can be used to restrict the search to only those 
enties belonging to that category. 

The `category` argment for your function should have a default value of `None`, 
in which case all entries are returned.



Required Signature:

    def get_restful_services(category):  
        pass  

For example, here are the first three items in the returned list, 
when the function was called with an argument `category='Music'`
(i.e. `https://api.publicapis.org/entries?category=Music`):

```
['https://docs.7digital.com/reference', 
'https://aimastering.com/api_docs/', 
'https://developer.apple.com/documentation/applemusicapi/']
```





````python
# submit (we will import your solution from this cell. Don't use Global Variables here.  )
# ------------------------------------------------------------------------------
def get_restful_services(category=None):
    import requests
    url = 'api.publicapis.org/entries'
    r = requests.get('https://api.publicapis.org/entries')
    data = r.json()
    #print(data)
    list = []
    x = data['entries']
    for item in x:
        if item['Category'] == category:
            link = item['Link']
            list.append(link)
        elif category == None:
            link = item['Link']
            list.append(link)
    return list

#get_restful_services('Jobs')


````
